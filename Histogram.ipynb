##BGR DAN RGB

buat syntax untuk membaca gambar yang telah pilih dari link url tertentu

disini karena kita menggunakan opencv, karena settingan default warnanya adalah bgr, maka kita ubah menjadi rgb agar sesuai

```python
#link image
url ="https://i.pinimg.com/564x/9a/44/7e/9a447e915e9d4fcd877141be100921fd.jpg"

image = io.imread(url)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
final_frame = cv.hconcat((image, image_2))
cv2_imshow(final_frame)
```

#HISTOGRAM BGR

plotting histogram dari gambar yang telah kita baca dari imread

```python
color = ('b','g','r')
for i,col in enumerate(color):
  histogr = cv.calcHist([image], [i], None, [256], [0,256])
  plt.plot(histogr, color = col)
  plt.xlim([0,256])
plt.show()
```

# GRAYSCALE IMAGE HISTOGRAM ANALYSIS

pengubahan colorspace gambar ke grayscale

```python
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

analisis histogram dari gambar grayscale

###bin 256

```python
plt.hist(gray_image.ravel(), bins = 256, range = [0,256])
plt.show()


print(gray_image.size)
print(gray_image.shape)
print(gray_image.dtype)
```

###bin 4

```python
plt.hist(gray_image.ravel(), bins = 4, range = [0,256])
plt.show()


print(gray_image.size)
print(gray_image.shape)
print(gray_image.dtype)
```

# IMAGE TRANSFORM-GRAYSCALE

##Inverse Grayscale
bright pixel became dark, dark became bright

```python
invert_gray = 255 - gray_image
cv2_imshow(invert_gray)
```

analisis histogram dari gambar grayscale yang telah di invert gelap-terangnya

```python
plt.hist(invert_gray.ravel(), bins = 256, range = [0,256])
plt.show()
```

## High Brightness Image
adding more bright, hazing-like image

analisis histogram dari gambar grayscale yang memiliki intensitas terang berlebih

```python
bright_image = (100.0/255)*gray_image + 100
cv2_imshow(bright_image)
```

```python
plt.hist(bright_image.ravel(), bins = 256, range = [0,256])
plt.show()
```

##Darken Image Transform
The Lightness level decreased

analisis histogram dari gambar grayscale yang telah diubah tingkat gelapnya ditambahkan

```python
darken_image = 255.0*(gray_image/255.0)**2
cv2_imshow(darken_image)
```

```python
plt.hist(darken_image.ravel(), bins = 256, range = [0,256])
plt.show()
```
